```r
data = read.csv("JFK.csv", header=TRUE)
head(data, n=5)
```

```r
data.ts = ts(data,start=c(1960,1),freq=12)
```

```r
par(mfrow=c(2,2))
plot(data.ts)
acf(data.ts, lag.max=12*4)

```

```r
## Estimate seasonality using ANOVA approach
month = season(data.ts)

## All seasonal mean effects (model without intercept)
model.ANOVA = lm(data.ts~month-1)
summary(model.ANOVA)

par(mfrow=c(2,1))
plot(data.ts)
lines(ts(fitted(model.ANOVA), start=c(1960,1), frequency=12), col="red")

MAPE=mean(abs(fitted(model.ANOVA)-data.ts)/data.ts) 
PM=sum((fitted(model.ANOVA)-data.ts)^2)/sum((data.ts-mean(data.ts))^2) 
MAPE
PM
```

```r
pre=predict(model.ANOVA, n.ahead=5)
pre
```

```r
## Estimate seasonality using cos-sin model
har2=harmonic(data.ts,2)
model.har=lm(data.ts~har2)
summary(model.har)

har.fit.res=fitted(model.har)-data.ts
har.fit.res.ts=ts(har.fit.res, start=c(1960,1), frequency=12)

par(mfrow=c(2,2))
plot(har.fit.res.ts, main="Harmonic residuals")
acf(har.fit.res.ts, main="harmonic residual")

MAPE=mean(abs(fitted(model.har)-data.ts)/data.ts) 
PM=sum((fitted(model.har)-data.ts)^2)/sum((data.ts-mean(data.ts))^2) 
MAPE
PM
```

```r
time.pts = c(1:length(data.ts))
time.pts = c(time.pts - min(time.pts))/max(time.pts)

x1 = time.pts
x2 = time.pts^2

lm.fit.har = lm(data.ts~x1+x2+har2)
summary(lm.fit.har)


par(mfrow=c(2,2))

resid=fitted(lm.fit.har)-data.ts
resid.ts=ts(resid, start=c(1960,1), frequency=12)

plot(resid.ts)
acf(resid.ts)

MAPE=mean(abs(fitted(lm.fit.har)-data.ts)/data.ts) 
PM=sum((fitted(lm.fit.har)-data.ts)^2)/sum((data.ts-mean(data.ts))^2) 
MAPE
PM
```

```r
lm.fit.ANOVA = lm(data.ts~x1+x2+month-1)
summary(lm.fit.ANOVA)

MAPE=mean(abs(fitted(lm.fit.ANOVA)-data.ts)/data.ts) 
PM=sum((fitted(lm.fit.ANOVA)-data.ts)^2)/sum((data.ts-mean(data.ts))^2) 
MAPE
PM
```

```r
modpred = predict(lm.fit.ANOVA, newdata=12
modpred
```

```r
data <- read.csv("AAPL.csv", header=TRUE)
data <- data[2]
train <- data[(1:313),]
test <- data[(314:318),]
train.ts <- ts(train,start=c(2014,1),freq=52)
```

```r
first.dif=diff(train.ts)
second.dif=diff(first.dif)

par(mfrow=c(2,2))
plot(first.dif, main="1st dif")
acf(first.dif, main="1st dif", lag.max=4*52)
pacf(first.dif, main="1st dif", lag.max=4*52)
plot(second.dif, main="2nd dif")
acf(second.dif, main="second dif", lag.max=4*52)
pacf(second.dif, main="second dif", lag.max=4*52)
```

```r
#ARIMA order selection template

test_modelA <- function(p,d,q){
 mod = arima(train.ts, order=c(p,d,q), method="ML")
 current.aic = AIC(mod)
 df = data.frame(p,d,q,current.aic)
 names(df) <- c("p","d","q","AIC")
 print(paste(p,d,q,current.aic,sep=" "))
 return(df)
 }
 
 orders = data.frame(Inf,Inf,Inf,Inf)
 names(orders) <- c("p","d","q","AIC")
 
 
 for (p in 0:4){
   for (d in 0:2){
     for (q in 0:4) {
       possibleError <- tryCatch(
         orders<-rbind(orders,test_modelA(p,d,q)),
         error=function(e) e
       )
       if(inherits(possibleError, "error")) next
 
     }
   }
 }
 orders <- orders[order(-orders$AIC),]
 tail(orders)
```

```r
modfit.1 = arima(train.ts, order = c(0,2,1), method = "ML") 
modfit.1
```

```r
modfit.2 = arima(train.ts, order = c(1,2,2), method = "ML") 
modfit.2
```

```r
#ARIMA root calculation template

#For model with order (p,d,q)

#MA Roots
abs(polyroot(c(1, coef(modfit.1))))
```

```r
abs(polyroot(c(1, coef(modfit.2)[1])))
abs(polyroot(c(1, coef(modfit.2)[2:3])))
```

```r

modpred.1 = predict(modfit.1,n.ahead=5) 

modpred

obs=test 
predict=modpred.1$pred 
MAPE=mean(abs(predict-obs)/obs) 
MSPE=mean((predict-obs)^2)
MAPE
MSPE
```

```r
modpred.2 = predict(modfit.2,n.ahead=5) 

modpred.2

obs=test 
predict=modpred.2$pred 
MAPE=mean(abs(predict-obs)/obs) 
 
MAPE

### Mean Squared Prediction Error (MSPE)
MSPE=mean((predict-obs)^2)
MSPE
### Mean Absolute Percentage Error (MAPE)
MAPE2=mean(abs(predict-obs)/obs)
MAPE2
```

```r


 
ubound = modpred.2$pred+1.96*modpred.2$se 

lbound = modpred.2$pred-1.96*modpred.2$se 
ymin = min(lbound) 
ymax = max(ubound) 

par(mfrow=c(2,2))
plot(1:5, test, type="l", ylim=c(ymin,ymax), xlab="Month from Jan 2020", ylab="stock", main="2nd model") 
points(1:5,modpred.2$pred,col="red") 
lines(1:5, ubound,lty=3,lwd= 2, col="blue")
lines(1:5, lbound,lty=3,lwd= 2, col="blue") 
 
```
